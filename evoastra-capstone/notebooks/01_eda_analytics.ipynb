```python
import pandas as pd

df = pd.read_csv("../data/raw/DataCoSupplyChainDataset.csv")

df.head()
```

```python
df.shape
df.columns
df.info()
df.isnull().sum().sort_values(ascending=False)
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="Type", data=df)
plt.title("Order Payment Type Distribution")
plt.show()
```

```python
sns.countplot(x="Delivery Status", data=df)
plt.xticks(rotation=45)
plt.title("Delivery Status Distribution")
plt.show()
```

```python
sns.histplot(df["Days for shipping (real)"], bins=30)
plt.title("Distribution of Shipping Days")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
df["Category Name"].value_counts().head(10).plot(kind="bar")
plt.title("Top Product Categories")
plt.show()
```

```python
df.to_csv("../data/processed/cleaned_supply_chain_data.csv", index=False)
```
