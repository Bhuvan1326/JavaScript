```python
import pandas as pd
import joblib

df = pd.read_csv("../data/processed/cleaned_supply_chain_data.csv")

model = joblib.load("../models/supply_chain_model.pkl")

df.head()
```

```python
X = df[[
    "Days for shipping (real)",
    "Days for shipment (scheduled)",
    "Benefit per order",
    "Sales per customer"
]]
```

```python
predictions = model.predict(X)

df["Predicted_Late_Delivery"] = predictions

df.head()
```
